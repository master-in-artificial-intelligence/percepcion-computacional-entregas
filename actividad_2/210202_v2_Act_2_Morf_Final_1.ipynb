## Introducción
Según la propia descripción de nuestra actividad, las operaciones espaciales tienen en cuenta las relaciones de vecindad entre los píxeles para aplicar diferentes tipos de mejoras a la imagen; mientras que las operaciones morfológicas, simplifican imágenes y conservan las principales características de los objetos.

En esta actividad, se deben analizar los filtros estudiados en clase y compararlos con otros que se estén usando actualmente para resolución de problemas reales. Es importante mostrar los resultados de los principales pasos e indicar ventajas e inconvenientes de cada uno.

Las transformaciones de imágenes vistas en clase se enmarcan dentro de tres grandes grupos:

1. Las operaciones puntuales, en las que el valor de un píxel de la imagen procesada, será función de su homólogo en la imagen original, sin tener en cuenta, en esta transformación, el valor de otros píxeles de la imagen.
  - Ajuste de intensidad.
  - Basadas en el histograma.
  - Operaciones aritméticas.
2. Filtros espaciales, en los que el píxel resultante es función del píxel correspondiente en la imagen original y de la vecindad de este, conocida como máscara, o kernel.
  - Filtros Paso-Bajo.
  - Filtros Paso-Alto.
  - Detección de Bordes.
3. Las operaciones morfológicas, suelen ser un postprocesado para facilitar otras operaciones, como la segmentación por ejemplo.
  - Dilatación.
  - Erosión.
  - Apertura.
  - Cierre.
  - Gradiente Morfológico.
  - Top-Hat.
  - Bottom-Hat.

Para detectar los bordes utilizaremos el algoritmo de Canny. El algoritmo de Canny se basa en tres grandes pasos: 
- Obtención del gradiente: en este paso se calcula la magnitud y orientación del vector gradiente en cada píxel. 
- Supresión no máxima: en este paso se logra el adelgazamiento del ancho de los bordes, obtenidos con el gradiente, hasta lograr bordes de un píxel de ancho.
- Histéresis  de  umbral:  en  este  paso  se  aplica  una  función  de  histéresis  basada  en  dos umbrales; con este proceso se pretende reducir la posibilidad de aparición de contornos falsos.

https://www.researchgate.net/publication/267240432_Deteccion_de_bordes_mediante_el_algoritmo_de_Canny

3. **Aplicación de perspectiva para selección del recorte** ("*4 Point OpenCV getPerspectiveTransform Example*").

Nos hemos ayudado de la siguiente web donde está el código explicado paso a paso:

https://www.pyimagesearch.com/2014/08/25/4-point-opencv-getperspective-transform-example/

```python
# import the necessary packages
import numpy as np
import argparse

# Aplicamos el cambio de perspectiva para quedarnos con la imagen recortada
# Código basado en: https://www.pyimagesearch.com/2014/08/25/4-point-opencv-getperspective-transform-example/
# Con varias modificaciones para permitir cambios de perspectiva de imágenes que no sean rectángulos perfectos

def length(v):
    return math.sqrt(v[0]**2+v[1]**2)
def dot_product(v,w):
    return v[0]*w[0]+v[1]*w[1]
def determinant(v,w):
    return v[0]*w[1]-v[1]*w[0]
def inner_angle(v,w):
    cosx=dot_product(v,w)/(length(v)*length(w))
    rad=math.acos(cosx) # in radians
    return rad*180/math.pi # returns degrees
def angle_clockwise(A, B):
    inner=inner_angle(A,B)
    det = determinant(A,B)
    if det<0: #this is a property of the det. If the det < 0 then B is clockwise of A
        return inner
    else: # if the det > 0 then A is immediately clockwise of B
        return 360-inner

def order_points(pts):
    # initialzie a list of coordinates that will be ordered
    # such that the first entry in the list is the top-left,
    # the second entry is the top-right, the third is the
    # bottom-right, and the fourth is the bottom-left
    rect = np.zeros((4, 2), dtype = "float32")
    # the top-left point will have the smallest sum
    s = pts.sum(axis = 1)
    rect[0] = pts[np.argmin(s)]
    ang=[]
    n=0
    #a partir de aquí vamos a calcular los ángulos entre el punto origen y cada uno de los otros trs puntos para ordenarlos en función del ángulo y saber cuál es el orden de los puntos
    for p in pts:
        if(rect[0,0] != p[0] or rect[0,1] != p[1]):
            ang.append(angle_clockwise(rect[0]-p,[0,1]))
            n+=1
        else:
            pts = np.delete(pts, n, axis=0)
            n+=1
    # print("Angulos:",ang)
    if(ang[0]<=ang[1] and ang[0]<=ang[2]): #si 0 es el punto más a la izquierda del punto inicial
        rect[1]=pts[0]
        if(ang[1]<=ang[2]):
            rect[2]=pts[1]
            rect[3]=pts[2]
        else:
            rect[2]=pts[2]
            rect[3]=pts[1]
    elif(ang[1]<=ang[0] and ang[1]<=ang[2]): # si 1 es el punto más a la izquierda del punto inicial
        rect[1]=pts[1]
        if(ang[0]<=ang[2]):
            rect[2]=pts[0]
            rect[3]=pts[2]
        else:
            rect[2]=pts[2]
            rect[3]=pts[0]
    else: # si 1 es el punto más a la izquierda del punto inicial
        rect[1]=pts[2]
        if(ang[0]<=ang[1]):
            rect[2]=pts[0]
            rect[3]=pts[1]
        else:
            rect[2]=pts[1]
            rect[3]=pts[0]
    return rect

def four_point_transform(image, pts):
	# obtain a consistent order of the points and unpack them
	# individually
	rect = order_points(pts)
	(tl, tr, br, bl) = rect
	# compute the width of the new image, which will be the
	# maximum distance between bottom-right and bottom-left
	# x-coordiates or the top-right and top-left x-coordinates
	widthA = np.sqrt(((br[0] - bl[0]) ** 2) + ((br[1] - bl[1]) ** 2))
	widthB = np.sqrt(((tr[0] - tl[0]) ** 2) + ((tr[1] - tl[1]) ** 2))
	maxWidth = max(int(widthA), int(widthB))
	# compute the height of the new image, which will be the
	# maximum distance between the top-right and bottom-right
	# y-coordinates or the top-left and bottom-left y-coordinates
	heightA = np.sqrt(((tr[0] - br[0]) ** 2) + ((tr[1] - br[1]) ** 2))
	heightB = np.sqrt(((tl[0] - bl[0]) ** 2) + ((tl[1] - bl[1]) ** 2))
	maxHeight = max(int(heightA), int(heightB))
	# now that we have the dimensions of the new image, construct
	# the set of destination points to obtain a "birds eye view",
	# (i.e. top-down view) of the image, again specifying points
	# in the top-left, top-right, bottom-right, and bottom-left
	# order
	dst = np.array([
		[0, 0],
		[maxWidth - 1, 0],
		[maxWidth - 1, maxHeight - 1],
		[0, maxHeight - 1]], dtype = "float32")
	# compute the perspective transform matrix and then apply it
	M = cv2.getPerspectiveTransform(rect, dst)
	warped = cv2.warpPerspective(image, M, (maxWidth, maxHeight))
	# return the warped image
	return warped
```

## Aplicación de Operaciones Morfológicas

En las operaciones morfológicas, el objetivo es transformar la estructura o forma de los objetos utilizando un elemento estructural. Estas operaciones pueden cambiar la forma y el tamaño de los objetos de la imagen para adaptarlo a nuestras necesidades.

Las operaciones morfológicas se pueden aplicar en imágenes binarias, en escala de grises y en color, en nuestro caso, trabajaremos con imágenes binarias. 

Hemos contemplado dos vías que difieren en el primer paso para la obtención de los objetos, o sus esbozos, a cuantificar.

### 1ª vía: Binarización y operaciones morfológicas.

En este priemra vía convertiremos las imágenes a las dos intensidades máxima (*foreground*) y mínima (*background*) a partir de cierto umbral para hacer una selección de nuestros objetos y así hacer un tratamiento de la imagen donde sólamente aparezca aquello que va ser estudiado. En este caso, el número de figuras relevantes de una carta.

Es muy común utilizar el algoritmo de Otsu para encontrar un buen umbral de diferenciación para los objetos (*foreground*), sobre un fondo (*background*). El método de Otsu elige un umbral que minimiza la varianza intraclase de los píxeles en blanco y negro con umbral

https://es.mathworks.com/help/images/ref/graythresh.html

En nuestro caso, el algoritmo de Otsu funciona bien, y es una herramienta totalmente válida, pero, ya que las operaciones morfológicas son un postprocesado de la imagen cuya función es la de preparar las imágenes para poder extraer de ellas ciertas *features*, se ha comprobado que un nivel cercano al obtenido con el método de Otsu, nos permite generalizar para todas las imágenes estudiadas, y, además, obtiene un resultado óptimo para la posterior cuantificación.

Veámos la diferencia de aplicar unos resultados u otros:

```python
# Pasamos a escala de grises, explicar por quéXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX
binary_1 = rgb2gray(lista_img_zoom[0])
binary_2 = rgb2gray(lista_img_zoom[1])
binary_3 = rgb2gray(lista_img_zoom[2])
binary_6 = rgb2gray(lista_img_zoom[3])
binary_7 = rgb2gray(lista_img_zoom[4])
binary_10= rgb2gray(lista_img_zoom[5])

# Obtención de la umbralización mediante el algoritmo de Otsu
thres1 = threshold_otsu(binary_1)
binary_1otsu = (binary_1 < thres1).astype(np.uint8)

thres2 = threshold_otsu(binary_2)
binary_2otsu = (binary_2 < thres2).astype(np.uint8)

thres3 = threshold_otsu(binary_3)
binary_3otsu = (binary_3 < thres2).astype(np.uint8)

thres6 = threshold_otsu(binary_6)
binary_6otsu = (binary_6 < thres2).astype(np.uint8)

thres7 = threshold_otsu(binary_7)
binary_7otsu = (binary_7 < thres2).astype(np.uint8)

thres10 = threshold_otsu(binary_10)
binary_10otsu = (binary_10 < thres2).astype(np.uint8)

thres_cte = 0.778
binary_1 = (binary_1 < thres_cte).astype(np.uint8)
binary_2 = (binary_2 < thres_cte).astype(np.uint8)
binary_3 = (binary_3 < thres_cte).astype(np.uint8)
binary_6 = (binary_6 < thres_cte).astype(np.uint8)
binary_7 = (binary_7 < thres_cte).astype(np.uint8)
binary_10 = (binary_10 < thres_cte).astype(np.uint8)

plt.figure(figsize=(15,10))

plt.subplot(261), plt.imshow(binary_1otsu)
plt.axis('off'),
plt.title('Nivel Umbral = %1.3f' %thres1, size=12)

plt.subplot(262), plt.imshow(binary_2otsu),
plt.axis('off')
plt.title('Nivel Umbral = %1.3f' %thres2, size=12)

plt.subplot(263), plt.imshow(binary_3otsu)
plt.axis('off'),
plt.title('Nivel Umbral = %1.3f' %thres3, size=12)

plt.subplot(264), plt.imshow(binary_6otsu),
plt.axis('off')
plt.title('Nivel Umbral = %1.3f' %thres6, size=12)

plt.subplot(265), plt.imshow(binary_7otsu)
plt.axis('off'),
plt.title('Nivel Umbral = %1.3f' %thres7, size=12)

plt.subplot(266), plt.imshow(binary_10otsu),
plt.axis('off')
plt.title('Nivel Umbral = %1.3f' %thres10, size=12)

plt.subplot(267), plt.imshow(binary_1),
plt.axis('off')
plt.title('Umbral = %1.3f' %thres_cte, size=12)

plt.subplot(268), plt.imshow(binary_2),
plt.axis('off')
plt.title('Umbral = %1.3f' %thres_cte, size=12)

plt.subplot(269), plt.imshow(binary_3),
plt.axis('off')
plt.title('Umbral = %1.3f' %thres_cte, size=12)

plt.subplot(2,6, 10), plt.imshow(binary_6),
plt.axis('off')
plt.title('Umbral = %1.3f' %thres_cte, size=12)

plt.subplot(2,6,11), plt.imshow(binary_7),
plt.axis('off')
plt.title('Umbral = %1.3f' %thres_cte, size=12)

plt.subplot(2,6,12), plt.imshow(binary_10),
plt.axis('off')
plt.title('Umbral = %1.3f' %thres_cte, size=12)
```

Otsu hace una umbralización buena para todas las imágenes, incluso deja fuera el brillo de la esquina inferior derecha que aparecen en el As y en el 10, pero sin embargo, imágenes como el 3 y el 6, van a necesitar de operaciones morfológicas mucho más "refinadas" que con el umbral seleccionado manualmente.

**¿Por qué ese valor de 0.778?.**

Porque para valores mayores empezamos a captar artefactos de brillos que son catalogados como *foreground*, y para valores menores tenenmos demasiados detalles de la imagen, acercándose más a la solución obtenida por el método de Otsu.

En el siguiente paso aplicaremos la **Apertura**. La apertura nos resultará útil para:

  -	Separar objetos que aparecen unidos, pero sin modificar las proporciones.
  -	Suaviza las esquinas cóncavas, no las convexas.
  -	También elimina pequeños ruidos que se encuentren en el background.


```python
kernelOp=[]
for im_zoom in lista_img_zoom:
    radio_erosion=round(0.03*max(im_zoom.shape[0],im_zoom.shape[1])) 
    kernelOp.append(np.ones((radio_erosion,radio_erosion)))

opening1 = cv2.morphologyEx(binary_1, cv2.MORPH_OPEN,kernelOp[0])
opening2 = cv2.morphologyEx(binary_2, cv2.MORPH_OPEN,kernelOp[1])
opening3 = cv2.morphologyEx(binary_3, cv2.MORPH_OPEN,kernelOp[2])
opening6 = cv2.morphologyEx(binary_6, cv2.MORPH_OPEN,kernelOp[3])
opening7 = cv2.morphologyEx(binary_7, cv2.MORPH_OPEN,kernelOp[4])
opening10 = cv2.morphologyEx(binary_10, cv2.MORPH_OPEN,kernelOp[5])

plt.figure(figsize=(12,12))

plt.subplot(231), plt.imshow(opening1)
plt.axis('off'),

plt.subplot(232), plt.imshow(opening2),
plt.axis('off')

plt.subplot(233), plt.imshow(opening3),
plt.axis('off')

plt.subplot(234), plt.imshow(opening6),
plt.axis('off')

plt.subplot(235), plt.imshow(opening7),
plt.axis('off')

plt.subplot(236), plt.imshow(opening10),
plt.axis('off')
```

Tras la Apertura anterior, la siguiente operación aplicada fué el **Cierre**. La idea era rellenar, o unir inconexos del *foreground*, pero no obtuvimos los resultados esperados a pesar de haber hecho diversas pruebas con el tamaño de elemento estructural (kernelCl), tanto de forma simétrica en ambos ejes como de forma asimétrica. 

En la mayoría de los casos llegamos al siguiente estado:

```python
kernelCl=[]
for im_zoom in lista_img_zoom:
    radio_erosion=round(0.10*max(im_zoom.shape[0],im_zoom.shape[1])) 
    kernelCl.append(np.ones((radio_erosion,radio_erosion)))
    
closing1 = cv2.morphologyEx(opening1, cv2.MORPH_CLOSE,kernelCl[0])
closing2 = cv2.morphologyEx(opening2, cv2.MORPH_CLOSE,kernelCl[1])
closing3 = cv2.morphologyEx(opening3, cv2.MORPH_CLOSE,kernelCl[2])
closing6 = cv2.morphologyEx(opening6, cv2.MORPH_CLOSE,kernelCl[3])
closing7 = cv2.morphologyEx(opening7, cv2.MORPH_CLOSE,kernelCl[4])
closing10 = cv2.morphologyEx(opening10, cv2.MORPH_CLOSE,kernelCl[5])

plt.figure(figsize=(12,12)) 

plt.subplot(231), plt.imshow(closing1)
plt.axis('off'),
plt.title('Apertura + cierre As de Diamantes', size=12)

plt.subplot(232), plt.imshow(closing2),
plt.axis('off')
plt.title('Apertura + cierre Dos de Picas', size=12)

plt.subplot(233), plt.imshow(closing3),
plt.axis('off')
plt.title('Apertura + cierre Tres de Oros', size=12)

plt.subplot(234), plt.imshow(closing6),
plt.axis('off')
plt.title('Apertura + cierre Seis de Bastos', size=12)

plt.subplot(235), plt.imshow(closing7),
plt.axis('off')
plt.title('Apertura + cierre Siete de Treboles', size=12)

plt.subplot(236), plt.imshow(closing10),
plt.axis('off')
plt.title('Apertura + cierre Diez de Diamantes', size=12)
```

Vemos que el Seis de Bastos ha quedado desfigurado, mientras que el Siete de Tréboles mantiene secciones sin unir.

```python
clos_list=[closing1,closing2,closing3,closing6,closing7,closing10]
n=1
for cls in clos_list:
    labeled,num = morph.label(cls, connectivity=2, return_num=True)
    print("La carta ", n, " es un ",num)
    n+=1
```

De lo anterior deducimos que el elemento estructural para el cierre es lo suficientemente grande para unir los bastos, y sin embargo no alcanza para unir los tréboles, luego este método no nos sirve.

En lugar del Cierre, que mantiene "demasiado" la forma original, como segunda transformación, haremos una **dilatación**, partiendo desde la apertura. 

```python
disk=[]
for im_zoom in lista_img_zoom:
    radio=round(0.03*max(im_zoom.shape[0],im_zoom.shape[1])) 
    disk.append(morph.disk(radius=radio))
    
dilated1 = morph.binary_dilation(opening1, selem=disk[0])
dilated2 = morph.binary_dilation(opening2, selem=disk[1])
dilated3 = morph.binary_dilation(opening3, selem=disk[2])
dilated6 = morph.binary_dilation(opening6, selem=disk[3])
dilated7 = morph.binary_dilation(opening7, selem=disk[4])
dilated10 = morph.binary_dilation(opening10, selem=disk[5])

plt.figure(figsize=(12,12)) 

plt.subplot(231), plt.imshow(dilated1)
plt.axis('off'),

plt.subplot(232), plt.imshow(dilated2),
plt.axis('off')

plt.subplot(233), plt.imshow(dilated3),
plt.axis('off')

plt.subplot(234), plt.imshow(dilated6),
plt.axis('off')

plt.subplot(235), plt.imshow(dilated7),
plt.axis('off')

plt.subplot(236), plt.imshow(dilated10),
plt.axis('off')
```

```python
dilat_list=[dilated1,dilated2,dilated3,dilated6,dilated7,dilated10]
n=1
for dils in dilat_list:
    labeled,num = morph.label(dils, connectivity=2, return_num=True)
    print("La carta ", n, " es un ",num)
    n+=1
```

El resultado es correcto. Podemos concluir que aplicando el proceso de apertura + dilatación a nuestras imágenes, conseguimos un mejor resultado que aplicando apertura + cierre. 

### Detección de bordes, relleno y binarización

En esta segunda vía variaremos el inicio, en lugar de aplicar la binarización mediante la obtención de un umbral, y a partir de ahí aplicar la apertura y dilatación, comenzaremos hallando los contornos externos de las imágenes binarizadas mediante la función de OpenCV *findContours*.

```python
list_binarias=[binary_1,binary_2,binary_3,binary_6,binary_7,binary_10]

lista_img_contornos=[]
lista_img_cont_rellenos=[]

for img in lista_img_zoom:
    lista_img_contornos.append(img.copy())
    lista_img_cont_rellenos.append(img.copy())

contornos=[]

#encontramos los contornos
for binaria in list_binarias:
    (cont, jerarquia) = cv2.findContours(binaria, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    contornos.append(cont)

n=0
#los dibujamos sobre la figura
for img_contornos in lista_img_contornos:
    cv2.drawContours(img_contornos, contornos[n], -1, (0, 255, 0), 3)
    n+=1

n=0
#Los rellenamos de negro para que contrasten en su totalidad con el fondo blanco
for img_contornos_rellenos in lista_img_cont_rellenos:
    cv2.fillPoly(img_contornos_rellenos, pts = contornos[n], color=(0,0,0))
    n+=1
    
plt.figure(figsize=(15,10))

plt.subplot(261), plt.imshow(lista_img_contornos[0])
plt.axis('off'),
plt.title('Bordes As de Diamantes', size=12)

plt.subplot(262), plt.imshow(lista_img_contornos[1]),
plt.axis('off')
plt.title('Bordes Dos de Picas', size=12)

plt.subplot(263), plt.imshow(lista_img_contornos[2])
plt.axis('off'),
plt.title('Bordes Tres de Oros', size=12)

plt.subplot(264), plt.imshow(lista_img_contornos[3]),
plt.axis('off')
plt.title('Bordes Seis de Bastos', size=12)

plt.subplot(265), plt.imshow(lista_img_contornos[4])
plt.axis('off'),
plt.title('Bordes Siete de Tréboles', size=12)

plt.subplot(266), plt.imshow(lista_img_contornos[5]),
plt.axis('off')
plt.title('Bordes Diez de Diamantes', size=12)

plt.subplot(267), plt.imshow(lista_img_cont_rellenos[0]),
plt.axis('off')
plt.title('Rellenos As de Diamantes', size=12)

plt.subplot(268), plt.imshow(lista_img_cont_rellenos[1]),
plt.axis('off')
plt.title('Rellenos Dos de Picas', size=12)

plt.subplot(269), plt.imshow(lista_img_cont_rellenos[2]),
plt.axis('off')
plt.title('Rellenos Tres de Oros', size=12)

plt.subplot(2,6, 10), plt.imshow(lista_img_cont_rellenos[3]),
plt.axis('off')
plt.title('Rellenos Seis de Bastos', size=12)

plt.subplot(2,6,11), plt.imshow(lista_img_cont_rellenos[4]),
plt.axis('off')
plt.title('Rellenos Siete de Tréboles', size=12)

plt.subplot(2,6,12), plt.imshow(lista_img_cont_rellenos[5]),
plt.axis('off')
plt.title('Rellenos Diez de Diamantes', size=12)
```

Una vez hemos rellenado los bordes vamos a eliminar el ruido mediante la operación de apertura.

```python
im_bin_rellenas=[]
open_rellena=[]
for img_contornos_rellenos in lista_img_cont_rellenos:
    im=rgb2gray(img_contornos_rellenos)
    thres = threshold_otsu(im)
    im2 = (im <= thres).astype(np.uint8)
    im_bin_rellenas.append(im2)

    radio_erosion=round(0.03*max(im2.shape[0],im2.shape[1])) #si rebajamos de 0.035 no se elimina completamente el ruido
    kernel = np.ones((radio_erosion,radio_erosion))
    open_rellena.append(cv2.morphologyEx(im2, cv2.MORPH_OPEN,kernel))


    
plt.figure(figsize=(15,10))

plt.subplot(261), plt.imshow(im_bin_rellenas[0])
plt.axis('off'),
plt.title('Binaria rell. As de Diamantes', size=12)

plt.subplot(262), plt.imshow(im_bin_rellenas[1]),
plt.axis('off')
plt.title('Binaria rell. Dos de Picas', size=12)

plt.subplot(263), plt.imshow(im_bin_rellenas[2])
plt.axis('off'),
plt.title('Binaria rell. Tres de Oros', size=12)

plt.subplot(264), plt.imshow(im_bin_rellenas[3]),
plt.axis('off')
plt.title('Binaria rell. Seis de Bastos', size=12)

plt.subplot(265), plt.imshow(im_bin_rellenas[4])
plt.axis('off'),
plt.title('Binaria rell. Siete de Tréboles', size=12)

plt.subplot(266), plt.imshow(im_bin_rellenas[5]),
plt.axis('off')
plt.title('Binaria rell. Diez de Diamantes', size=12)

plt.subplot(267), plt.imshow(open_rellena[0]),
plt.axis('off')
plt.title('Apertura rell. As de Diamantes', size=12)

plt.subplot(268), plt.imshow(open_rellena[1]),
plt.axis('off')
plt.title('Apertura rell. Dos de Picas', size=12)

plt.subplot(269), plt.imshow(open_rellena[2]),
plt.axis('off')
plt.title('Apertura rell. Tres de Oros', size=12)

plt.subplot(2,6, 10), plt.imshow(open_rellena[3]),
plt.axis('off')
plt.title('Apertura rell. Seis de Bastos', size=12)

plt.subplot(2,6,11), plt.imshow(open_rellena[4]),
plt.axis('off')
plt.title('Apertura rell. Siete de Tréboles', size=12)

plt.subplot(2,6,12), plt.imshow(open_rellena[5]),
plt.axis('off')
plt.title('Apertura rell. Diez de Diamantes', size=12)
```

El último paso será el mismo que en la primera vía, la operación de **dilatación** para expandir el *foreground* uniendo zonas inconexas como los tréboles del 7, ya que necesitamos que un trebol sea un único objeto para hacer correctamente la cuantificación de los objetos.

```python

disk2=[]
for im_zoom in lista_img_zoom:
    radio=round(0.03*max(im_zoom.shape[0],im_zoom.shape[1]))
    disk2.append(morph.disk(radius=radio))
    
dilated1_r = morph.binary_dilation(open_rellena[0], selem=disk2[0])
dilated2_r = morph.binary_dilation(open_rellena[1], selem=disk2[1])
dilated3_r = morph.binary_dilation(open_rellena[2], selem=disk2[2])
dilated6_r = morph.binary_dilation(open_rellena[3], selem=disk2[3])
dilated7_r = morph.binary_dilation(open_rellena[4], selem=disk2[4])
dilated10_r = morph.binary_dilation(open_rellena[5], selem=disk[5])

plt.figure(figsize=(12,12)) 

plt.subplot(231), plt.imshow(dilated1_r)
plt.axis('off'),

plt.subplot(232), plt.imshow(dilated2_r),
plt.axis('off')

plt.subplot(233), plt.imshow(dilated3_r),
plt.axis('off')

plt.subplot(234), plt.imshow(dilated6_r),
plt.axis('off')

plt.subplot(235), plt.imshow(dilated7_r),
plt.axis('off')

plt.subplot(236), plt.imshow(dilated10_r),
plt.axis('off')
```

## CONCLUSIONES

La binarización de la imagen para su posterior tratamiento, bien sea el hecho en esta práctica, o cualquier otro, resulta un proceso nada trivial. Se ha comprobado experimentalmente que la generalización del método de Otsu, método bien conocido y ampliamente eplicado, no ha resultado ser suficiente para obtener un nivel de umbralización adecuado para simplificar los trabajos posteriores sobre las imágenes. Sin embargo obtiene muy buenos resultados como se ha podido comprobar, y nos ha servido de base para tomaar un valor idóneo para nuestro caso particular.

La erosión y la dilatación son los operadores morfológicos esenciales. A partir de ellos se crean el resto de operadores. 

En nuestro caso, los operadores en los que nos hemos basado para la obtención de los resultados, han sido:
- 1º) La apertura, consistente en una erosión seguida de una dilatación, para eliminar aquellas partes de la imagen de menor tamaño, o incluso ruido.
- 2º) Tras esa primera etapa, la operación básica de dilatación ha arrojado muy buenos resultados, expandiendo contornos hasta fusionarlos para obtener los objetos deseados.

Computacionalmente hablando, las operaciones básicas de erosión y dilatación son más eficientes que la apertura y cierre, entre otras, con lo que siempre que la solución sea buena, nuestro código será más eficiente con este tipo de operaciones.
